# 📊 02 — Exploratory Data Analysis
India Traffic Accident Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from pathlib import Path

plt.style.use('dark_background')
sns.set_palette('husl')

ROOT = Path('.').resolve().parent
df = pd.read_csv(ROOT / 'data/processed/accidents_clean.csv', parse_dates=['datetime'])
print(f'Shape: {df.shape}')
df.head()
```

```python
# Basic stats
print('=== Dataset Overview ===')
print(f'Records     : {len(df):,}')
print(f'Date range  : {df["datetime"].min().date()} → {df["datetime"].max().date()}')
print(f'States      : {df["state"].nunique()}')
print(f'Districts   : {df["district"].nunique()}')
print(f'Total deaths: {df["fatalities"].sum():,}')
print(f'Fatality %  : {df["fatalities"].sum()/len(df)*100:.2f}%')
df.describe()
```

```python
# Missing value check
missing = df.isnull().sum()
missing[missing > 0]
```

```python
# Distribution of fatalities
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

df['fatalities'].value_counts().head(10).plot(kind='bar', ax=axes[0], color='#e94560')
axes[0].set_title('Fatality Count Distribution')
axes[0].set_xlabel('Number of Fatalities per Accident')

df['severity'].value_counts().plot(kind='pie', ax=axes[1], autopct='%1.1f%%',
                                    colors=['#e94560','#f5a623','#4ecdc4'])
axes[1].set_title('Severity Distribution')
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
num_cols = ['fatalities','injuries','total_casualties','num_vehicles','hour','month','is_weekend']
corr = df[num_cols].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdYlGn', center=0,
            square=True, linewidths=0.5)
plt.title('Correlation Matrix — Numerical Features')
plt.tight_layout()
plt.show()
```

```python
# State-wise accident count
state_counts = df['state'].value_counts().head(12)
fig = px.bar(state_counts.reset_index(), x='count', y='state',
             orientation='h', title='Top 12 States by Accident Count',
             color='count', color_continuous_scale='Reds',
             template='plotly_dark')
fig.update_layout(yaxis=dict(autorange='reversed'))
fig.show()
```
